```python
#importing the datasource
import pandas as pd
df = pd.read_excel("RSC.xlsx")
```

```python
#1: Deleting the empty columns i.e. "Status and"
del df['Status']
del df['unnamed1']
```

```python
#2: Replacing values on "Gender"
df['Gender'] = df['Gender'].str.replace('Malealeale','Male')
df['Gender'] = df['Gender'].str.replace('Femaleemaleemale','Female')
```

```python
#3: Replacing the values on "Marital_Status"
df['Marital_Status'] = df['Marital_Status'].astype(str)
df['Marital_Status'] = df['Marital_Status'].str.replace('1','Married')
df['Marital_Status'] = df['Marital_Status'].str.replace('0','Single')
```

```python
#4: Replacing the empty cells in "Amount" with '0'
df['Amount'] = df['Amount'].fillna(0)
```

```python
#assigning the cleaned df to dff
dff = df
```

```python
#5: Calculating the total number of Customers - unique/distinct count
dff['Cust_name'].nunique()
```

```python
#6: Calculating the sum of orders by occupation with grand-total
dff.pivot_table(index = "Occupation",
               values = "Orders",
               aggfunc = "sum",
            margins = True,
            margins_name = 'Grand Total')
```

```python
#6: Calculating the sum of orders by occupation
ordoccup = dff.pivot_table(index = "Occupation",
               values = "Orders",
               aggfunc = "sum")

ordoccup.sort_values(by=['Orders'], ascending = False)
```

```python
#sorting the values by "Occupation"
ordoccpvt = ordoccup.sort_values('Occupation', ascending = True)
ordoccpvt
```

```python
#7: Top 10 Product_Category by Sum of Amount
prodamt = dff.pivot_table(index = "Product_Category",
               values = "Amount",
               aggfunc = "sum")

prodamt.sort_values(by=['Amount'], ascending = False).head(10)
```

```python
#7: Top 10 Product_Category by Sum of Amount with percentage of total
prodamt = dff.pivot_table(index = "Product_Category",
               values = "Amount",
               aggfunc = "sum")

prodamtpvt = prodamt.sort_values(by=['Amount'], ascending = False).head(10)

prodamtpvt["%"] = 100*prodamtpvt['Amount'] / prodamtpvt['Amount'].sum()
prodamtpvt.round(2)

```

```python
#8: Orders Summary by Gender and Age Group
ordgenage = dff.pivot_table(index = ['Gender','Age'],
               values = "Orders",
               aggfunc = "sum")

ordgenagepvt = ordgenage.sort_values(by=['Orders'], ascending = False).head(10)

ordgenagepvt["%"] = 100*ordgenagepvt['Orders'] / ordgenagepvt['Orders'].sum()
ordgenagepvt.round(2)
```
