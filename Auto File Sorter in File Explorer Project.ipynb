# Auto File Sorter in File Explorer with AlexTheAnalyst Tutorial

```python
import os, shutil
```

```python
path = r"C:/Users/jesul/Downloads/"
```

```python
file_name = os.listdir(path)
```

```python
#creating folders for sorting
#checking first
folder_names = ['csv files', 'image files', 'text files', 'setup files', 'video files', 'zip files']

for loop in range(0,6):
    if not os.path.exists(path + folder_names[loop]):
        print((path + folder_names[loop]))
        os.makedirs((path + folder_names[loop]))

for file in file_name:
    if ".csv" in file and not os.path.exists(path + "csv files/" + file):
        shutil.move(path + file, path + "csv files/" + file)
    elif ".png" in file and not os.path.exists(path + "image files/" + file):
        shutil.move(path + file, path + "image files/" + file)
    elif ".jpg" in file and not os.path.exists(path + "image files/" + file):
        shutil.move(path + file, path + "image files/" + file)
    elif ".jpeg" in file and not os.path.exists(path + "image files/" + file):
        shutil.move(path + file, path + "image files/" + file)
    elif ".gif" in file and not os.path.exists(path + "image files/" + file):
        shutil.move(path + file, path + "image files/" + file)
    elif ".pdf" in file and not os.path.exists(path + "text files/" + file):
        shutil.move(path + file, path + "text files/" + file)
    elif ".txt" in file and not os.path.exists(path + "text files/" + file):
        shutil.move(path + file, path + "text files/" + file)
    elif ".docx" in file and not os.path.exists(path + "text files/" + file):
        shutil.move(path + file, path + "text files/" + file)
    elif ".pptx" in file and not os.path.exists(path + "text files/" + file):
        shutil.move(path + file, path + "text files/" + file)
    elif ".exe" in file and not os.path.exists(path + "setup files/" + file):
        shutil.move(path + file, path + "setup files/" + file)
    elif ".mp4" in file and not os.path.exists(path + "video files/" + file):
        shutil.move(path + file, path + "video files/" + file )
    elif ".zip" in file and not os.path.exists(path + "zip files/" + file):
        shutil.move(path + file, path + "zip files/" + file )
    else:
        print("That's All")
```
